### Split de datos entre entrenamiento y test

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(digits.data, #Datos
                                                   digits.target,#Variable objetivo
                                                   test_size=0.2, #Relación 80/20
                                                   random_state=0)
```

```python
model = LogisticRegression() #Modelo
model.fit(x_train,y_train) #fit
```

```python
prediction = model.predict(x_test) #Predicción desde el x_test habiendo entrenado
prediction
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test,prediction) 
cm
```

* Evaluar valores NA

```python
import os
```

```python
os.getcwd()
os.chdir('Datos')
```

```python
churn = pd.read_csv('churn.csv')
churn.head(5)
```

```python
churn.TotalCharges = pd.to_numeric(churn.TotalCharges, 
                                  errors='coerce')
```

```python
churn.info()
```

```python
churn.isnull().sum()
```

```python
churn.dropna(inplace=True)
```

```python
churn.drop(columns='customerID',axis=1,inplace=True)
churn
```

```python
churn['Churn'].replace(to_replace='Yes',value=1, inplace=True)
churn['Churn'].replace(to_replace='No',value=0, inplace=True)
churn.Churn
```

```python
df = churn.copy()
```

```python
df.head(5)
```

```python
df = pd.get_dummies(df)
df.head(5)
```

```python
fig = plt.figure(figsize=(9,9))
df.corr()['Churn'].sort_values(ascending=True).plot(kind='bar')
```

#### Escalabilidad

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df)
df_scaled = pd.DataFrame(df_scaled)
df_scaled.columns = df.columns
df_scaled.head(5)
```

```python
sns.countplot(data=churn,x='gender',hue='Churn')
```

```python
def plot_categorical(column):
    fig = plt.figure(figsize=(7,7))
    sns.countplot(data=churn,x=column,hue='Churn')
    plt.show()
```

```python
column_cat = churn.select_dtypes(include='object').columns
```

```python
for i in column_cat:
    plot_categorical(i)
```

```python
sns.pairplot(data = churn, hue='Churn')
```

### Regresión Logística Binomial

```python
X = df_scaled.drop('Churn',axis=1)
y = df_scaled['Churn'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                   test_size=0.3,
                                                   random_state=42)
```

```python
model = LogisticRegression()
result = model.fit(X_train,y_train)
```

```python
prediction_test = model.predict(X_test)
```

```python
print(metrics.accuracy_score(y_test,prediction_test))
```

### Anotaciones con respecto a los resultados

#### El Churn es más probable que ocurra bajo las siguientes condiciones:
* Si la tasa total es alta

#### Contrariamente es más probable que se quede bajo las siguientes condiciones:

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
fig = plt.figure(figsize=(9,9))
cm = confusion_matrix(y_test,prediction_test,
                     labels=model.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, 
                             display_labels=model.classes_)
disp.plot(cmap='gray')
plt.show()
```
